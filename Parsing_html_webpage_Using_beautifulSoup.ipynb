```python
# Base code for all exercises to avoid retyping the same thing

import requests
from bs4 import BeautifulSoup

url = 'https://www.python.org/'

reqs = requests.get(url)
content = reqs.text
soup = BeautifulSoup(content, 'html.parser')
```

```python
# 1. Find the title tags from a given html document.

print(soup.find("title"))
```

```python
# 2. Get the number of paragraph tags of a given html document.

print(len(soup.find_all("p")))
```

```python
# 3. Extract the text in the first paragraph tag of a given html document.

print(soup.find("p").text)
```

```python
#4. Extract all the URLs from the webpage python.org that are nested within <li> tags
 
urls = []

for x in soup.findAll('li'):
    a = x.find('a')
    if 'href' in a.attrs:
        url = a.get('href')
        urls.append(url)
    else:
        pass

for url in urls:
    print(url)
```

```python
#5. Extract all the text from a given web page

print(soup.text)
```

```python
#6. Retrieve the HTML code of the title, its text, and the HTML code of its parent.

print(soup.title)
print(soup.title.text)
print(soup.title.parent)
```

```python
#7. Create a Beautiful Soup parse tree into a nicely formatted Unicode string, 
#   with a separate line for each HTML/XML tag and string.

print(soup.prettify())
```

```python
#8. Remove the contents of a tag in a given html document.

print("Before:")
print(soup.a) #before
tag = soup.a
tag = tag.clear()
print("\nAfter:")
print(soup.a) #after
```

```python
#9. Wrap an element in the specified tag and create the new wrapper.

print("Before:")
print(soup.p.wrap(soup.new_tag("tr"))) #before
print("\nAfter:")
print(soup.p.wrap(soup.new_tag("div"))) #after
```

```python
#10. Remove a tag from a given tree of html document and destroy it and its contents.

a_tag = soup.a 
print("Before:")
print(a_tag) #before

new_tag = soup.a.decompose()
print("\nAfter:")
print(new_tag) #after
```
